### Setting up intermediary exploratory tables into a spreadsheet

In this step, I am trying write intermidary findings into different sheets within a spreadsheet named Navaneeth_Data_Observations.xlsx created within my GitHub Repository

```python
import pandas as pd
import os

# Step 1: Read dataset directly from GitHub ZIP
url = 'https://github.com/nmadavan/APEX-PROJECT-I/raw/main/online_retail.csv.zip'
df = pd.read_csv(url, compression='zip', encoding='unicode_escape')

# Step 2: Create summary of unique InvoiceNo and their counts
invoice_summary = (
    df['InvoiceNo']
    .value_counts()                # Count occurrences
    .reset_index()                 # Convert to DataFrame
    .rename(columns={'index': 'InvoiceNo', 'InvoiceNo': 'Count'})  # Rename columns
    .sort_values(by='Count', ascending=True)  # Sort ascending by count
)

# Step 3: Write summary to Excel
excel_path = '/content/Navaneeth_Data_Observations.xlsx'

with pd.ExcelWriter(excel_path, engine='openpyxl', mode='a' if os.path.exists(excel_path) else 'w') as writer:
    invoice_summary.to_excel(writer, sheet_name='Unique Invoices', index=False)

print("✅ 'Unique Invoices' sheet written successfully!")
print(invoice_summary.head())
```
